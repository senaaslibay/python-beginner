```
import time
import pandas as pd
import numpy as np


altın=[]
dolar=[]
euro=[]

def donustur(veriler,a,b,c):
  sonuc=(veriler[a].loc[b,"Son.1"])/c
  return sonuc


def hesapla(x):
  sonuc=99000/x
  return sonuc

def yazdır(dizi,string):
  for i in dizi:
    print(str(i)+" kadar "+string+" alabilirsiniz.\n")


for i in range(60):
  
  veriler=pd.read_html('http://finans.mynet.com/')
  
  altın_d=donustur(veriler,11,2,100)
  dolar_d=donustur(veriler,9,0,10000)
  euro_d=donustur(veriler,9,1,10000)
  
  _altın=hesapla(altın_d)
  _dolar=hesapla(dolar_d)
  _euro=hesapla(euro_d)
  
  altın.append(_altın)
  dolar.append(_dolar)
  euro.append(_euro)
  
  time.sleep(60)
  
 
yazdır(altın,"çeyrek altın")
yazdır(dolar,"dolar")
yazdır(euro,"euro")

  
"""print("Çeyrek altın: " + str(altın_d) + " TL")
print("99.000 TL'ye "+ str(_altın)+" çeyrek altın alabilirsiniz.\n")

print("Dolar: " + str(dolar_d)+ " TL")
print("99.000 TL'ye "+ str(_dolar)+" dolar alabilirsiniz.\n")

print("Euro :" + str(euro_d)+ " TL")
print("99.000 TL'ye "+ str(_euro)+" euro alabilirsiniz.\n")
  
 """ 
  
    
```
